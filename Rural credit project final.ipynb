```python
occ_loan = (
    df.groupby('occupants_count')['loan_amount']
      .sum()
      .sort_values(ascending=False)
      .head(5)
)

x = np.arange(len(occ_loan.index))

plt.figure(figsize=(9,6))

bars = plt.bar(
    x,
    occ_loan.values,
    color='lightsteelblue',
    edgecolor='black'
)

plt.xlabel('Occupants Count')
plt.ylabel('Total Loan Amount')
plt.title('Top 5 Occupants Count vs Total Loan Amount')
plt.xticks(x, occ_loan.index)


plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
df['home_ownership'] = df['home_ownership'].replace({
    1.0: 'Owned',
    0.0: 'Rented'
})

top5_business = df['primary_business'].value_counts().head(5).index

comparison_data = (
    df[df['primary_business'].isin(top5_business)]
    .groupby(['home_ownership', 'primary_business'])
    .size()
    .unstack(fill_value=0)
)

colors = ['#A9CCE3', '#D6EAF8', '#A3E4D7', '#D5F5E3', '#EBDEF0']

ax = comparison_data.plot(
    kind='bar',
    figsize=(9,6),
    color=colors,
    edgecolor='black'
)

plt.xlabel('Home Ownership')
plt.ylabel('Number of Respondents')
plt.title('Home Ownership vs Top 5 Primary Businesses')
plt.xticks(rotation=0)

plt.legend(
    title='Primary Business',
    bbox_to_anchor=(1.05, 1),
    loc='upper left'
)


for bar in ax.patches:
    ax.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height(),
        int(bar.get_height()),
        ha='center',
        va='bottom',
        fontsize=9
    )

plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
dep_data = (
    df.groupby('occupants_count')[['old_dependents', 'young_dependents']]
      .sum()
      .sort_index()
)

old_color = 'burlywood'   
young_color = 'lemonchiffon'  

x = np.arange(len(dep_data.index))
width = 0.35

plt.figure(figsize=(10,6))

bars1 = plt.bar(
    x - width/2,
    dep_data['old_dependents'],
    width,
    label='Old Dependents',
    color=old_color,
    edgecolor='black'
)

bars2 = plt.bar(
    x + width/2,
    dep_data['young_dependents'],
    width,
    label='Young Dependents',
    color=young_color,
    edgecolor='black'
)

plt.xlabel('Occupants Count')
plt.ylabel('Number of Dependents')
plt.title('Occupants Count vs Old and Young Dependents')
plt.xticks(x, dep_data.index)
plt.legend()

# Value labels
for bars in [bars1, bars2]:
    for bar in bars:
        plt.text(
            bar.get_x() + bar.get_width()/2,
            bar.get_height(),
            int(bar.get_height()),
            ha='center',
            va='bottom',
            fontsize=9
        )

plt.grid(axis='y', linestyle='--', alpha=0.4)

# 🔹 ADD THIS LINE TO REMOVE THOUSANDS FORMAT
plt.ticklabel_format(style='plain', axis='y')

plt.tight_layout()
plt.show()
```
